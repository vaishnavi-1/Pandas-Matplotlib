```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("sales_train.csv" )
df = data.head(20000)
```

```python
df
```

## Plot the graphs using different attributes of the dataset

```python
plt.bar(x = df["shop_id"], height = df["item_id"])
plt.xlabel("Shop id")
plt.ylabel("Item id")
```

```python
plt.figure(figsize=(7,4))
plt.bar(x=df['shop_id'],height=df['item_price'])
plt.xlabel('Shop id')
plt.ylabel('Item Price')
```

```python
x = df.shop_id
y = df.item_id
plt.scatter(x,y)
plt.show()
```

```python
x = df.shop_id
y = df.item_price
plt.scatter(x,y)
plt.show()
```

```python
pt = sns.histplot(df['item_id'])
```

```python
at = sns.histplot(df['shop_id'])
```

```python
df.boxplot(by ='shop_id', column =['item_id'], grid = False)
```

```python
df.boxplot(by ='shop_id', column =['item_price'], grid = False)
```

## Plot the Heatmap

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(df.corr(), center=0, cmap='BuGn')
ax.set_title('Heatmap')
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(), center=0, cmap='Accent_r', annot=True)
```
