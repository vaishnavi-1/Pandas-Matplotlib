# Types Of Plots

## Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
x=[2,13,9,45]
y=[34,26,17,8]
plt.plot(x,y)
plt.show()
```

```python
x = [11,12,23] 
y = [22,14,36] 

plt.plot(x, y, label = "line1") 

x1 = [14,20,3] 
y1 = [40,8,33] 

plt.plot(x1, y1, label = "line2") 

plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Two lines on one graph') 

plt.legend() 

plt.show()
```

```python
x = [11,12,23,4,51,46] 
y = [22,14,36,27,38,9]

plt.plot(x,y, color='brown', linestyle='dashed', linewidth= 4, marker ='o' , markerfacecolor='blue', markersize =10)

plt.ylim(1,60)
plt.xlim(1,60)

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

## Bar Plot

```python
import matplotlib.pyplot as plt 

x = [10, 22, 13, 34, 30] 
 
h = [19, 28, 36, 24, 52] 

tick_label = ['one', 'two', 'three', 'four', 'five'] 

plt.bar(x, h, tick_label = tick_label, width = 0.7, color = ['yellow', 'red']) 
 
plt.xlabel('x-axis') 

plt.ylabel('y-axis') 

plt.title('Bar graph') 

plt.show()
```

## Histogram

```python
x = [34,54,5,25,87,6]
plt.hist(x)
plt.show()
```

```python
item_id = [2,44,31,20,50,36,87,94,49,57,6,20,74,36,20,36,2,8,90,92,36,74,72]
range = (0,100)
bins = 10

plt.hist(item_id, bins, range, color ="orange", histtype = "bar", rwidth = 0.7)
plt.ylabel("Item id")
plt.xlabel("Item Price")

plt.show()
```

## Scatter Plot

```python
x =[23,13,7,4,30,41,56,39,72]
y = [67,89,12,33,44,21,78,68,93]
plt.scatter(x,y)
plt.show()
```

```python
x = [23,13,7,4,30,41,56,39,72,40,58,60,31,3]
y = [67,89,12,33,44,21,78,68,93,17,23,22,2,4]
plt.scatter(x,y, label ="stars", color ="green", marker="*",s=30)
plt.show()
```

## Pie-Chart

```python
company = ['Jharkhand','Orissa','West Bengal','Bangalore']

s = [9,6,2,7] 

colors = [ 'y', 'w', 'm','r'] 

plt.pie(s, labels = company, colors=colors, startangle=90, shadow = True, explode = (0.2, 0, 0, 0), radius = 1.6, autopct = '%1.1f%%') 

plt.legend() 

plt.show()
```

## Plotting curves of given equation

```python
x = np.arange(0, 10*(np.pi), 0.7) 
y = np.sin(x) 

plt.plot(x, y) 

plt.show()
```

```python
x = np.arange(0, 10*(np.pi), 0.52) 
y = np.cos(x) 

plt.plot(x, y) 

plt.show()
```

```python
x = np.arange(0, 10*(np.pi), 0.52) 
y = np.tan(x) 

plt.plot(x, y) 

plt.show()
```

## Contour Graph

```python
xlist = np.linspace(-8, 8.0, 50)
ylist = np.linspace(-15.0,35.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp)                    # Add a colorbar to a plot

ax.set_title('Filled Contours Plot')

#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```

## Box Plot

```python
np.random.seed(20)
data = np.random.normal(80,60,100)
plt.boxplot(data)
plt.show()
```

```python
np.random.seed(20) 
data_1 = np.random.normal(10, 30, 500) 
data_2 = np.random.normal(90, 40, 500) 
data_3 = np.random.normal(80, 50, 500) 
data_4 = np.random.normal(20, 10, 500) 
data = [data_1, data_2, data_3, data_4] 

fg = plt.figure(figsize =(9, 6)) 

axis = fg.add_axes([0, 0, 1, 1]) 
 
bp = axis.boxplot(data) 

plt.show()
```
